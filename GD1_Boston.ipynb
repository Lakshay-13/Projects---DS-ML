```python
import numpy as np
import pandas as pd
from sklearn import preprocessing
from sklearn import datasets
def step_gradient(matrix, learning_rate, current_coeff):
    N = len(matrix)
    M = len(current_coeff)
    coeff_slope = [0 for i in range(M)]
    new_coeff = []
    
    for i in range(N):
        
        x = matrix[:,:105]
        y = matrix[i,105]
        value = y         
        for j in range(M):
            value = value - current_coeff[j]*x[i][j]
        for j in range(M):
            coeff_slope[j] += -(2/N) * value * matrix[i][j]
    
    for i in range(M):
        new_coeff.append(current_coeff[i] - learning_rate*coeff_slope[i])
    return new_coeff
def cost_gd(coeff, matrix):
    M = len(coeff)
    total_cost = 0
    
    
    for i in range(len(matrix)):
        x = matrix[:,:105]
        y = matrix[i,105]
        value = y 
        for j in range(M):
            value = value - coeff[j]*x[i][j]
        total_cost += value**2
    return total_cost/len(matrix)
def gd_runner(matrix, learning_rate, num_iterations):
    # b = 0
    # m = 0
    cost_list = []
    coeff = [0 for i in range(105)]
    
    cost_gd(coeff,matrix)
    
    for i in range(num_iterations):
        cur_cost = cost_gd(coeff,matrix)
        cost_list.append(cur_cost)
        coeff = step_gradient(matrix, learning_rate,coeff)
    cost_list.append(cost_gd(coeff,matrix))
    return (coeff,cost_list)
def predict(data_test,final_coeff,c):
    scaler=preprocessing.StandardScaler()
    scaler.fit(data_test)
    scaler.transform(data_test)
    df=pd.DataFrame(data_test)
    boston = datasets.load_boston()
    df.columns=boston.feature_names
    n=len(df.columns)
    for i in range(n):
        for j in range(i,n):
            r=df.columns[i]+'_'+df.columns[j]
            df[r]=df[df.columns[i]]*df[df.columns[j]]
    df[104]=1
    data_test=df.values
    return (np.dot(data_test,final_coeff))
def run():
   
    boston = datasets.load_boston()
    """df = pd.DataFrame(boston.data)
    df.columns = boston.feature_names"""
    data=np.genfromtxt('0000000000002417_training_boston_x_y_train.csv',delimiter=',')
    df=pd.DataFrame(data[:,:-1])
    df.columns=boston.feature_names
    n=len(df.columns)
    for i in range(n):
        for j in range(i,n):
            r=df.columns[i]+'_'+df.columns[j]
            df[r]=df[df.columns[i]]*df[df.columns[j]]
    one_matrix = [1 for i in range(len(df))]
    df['LAST_COEFF'] = one_matrix
    df['PRICE'] = data[:,-1]
    matrix = df.as_matrix(columns = df.columns[:])
    scaler=preprocessing.StandardScaler()
    scaler.fit(matrix)
    scaler.transform(matrix)
    learning_rate = 0.01
    num_iterations = 3200
    N = len(matrix)
    final_coeff,cost_list = gd_runner(matrix, learning_rate, num_iterations) 
    data_test=np.genfromtxt('0000000000002417_test_boston_x_test.csv',delimiter=',')
    y_pred=predict(data_test,final_coeff,cost_list[-1])
    np.savetxt('Gradient Descent 1.csv',y_pred,fmt='%1.8f')
run()
```
